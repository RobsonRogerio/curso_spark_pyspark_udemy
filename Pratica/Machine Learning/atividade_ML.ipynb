```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("TesteNotebook").getOrCreate()
```

```python
from pyspark.ml.feature import RFormula
from pyspark.ml.classification import NaiveBayes
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
iris = spark.read.csv("/home/robson/curso_spark_pyspark_udemy/iris.csv", header=True, inferSchema=True, sep=",")
```

```python
iris.show()
```

```python
formula = RFormula(formula="class ~ .", featuresCol="features", labelCol="label", handleInvalid="skip")
```

```python
iris_trans = formula.fit(iris).transform(iris).select("features", "label")
```

```python
iris_trans.show(150)
```

```python
iris_treino, iris_teste = iris_trans.randomSplit([0.7, 0.3], seed=42)
```

```python
iris_treino.count(), iris_teste.count()
```

```python
nb = NaiveBayes(labelCol="label", featuresCol="features")
```

```python
modelo = nb.fit(iris_treino)
```

```python
previsao = modelo.transform(iris_teste)
```

```python
previsao.show(truncate=False)
```

```python
avaliar = MulticlassClassificationEvaluator(predictionCol="prediction", labelCol="label", metricName="accuracy")
```

```python
resultado = avaliar.evaluate(previsao)
```
